## Task 5: Perceptual Pitfalls

**Instructions:**
- Create two bar plots of average price by cut: one with a truncated y-axis and one with a zero-based y-axis.

**Reflection:**
- How does the y-axis choice affect your perception of the data? Why is this important for ethical data communication?


```python
import matplotlib.pyplot as plt
import seaborn as sns

diamonds = sns.load_dataset('diamonds')

# Task 5: Your code here
fig, axs = plt.subplots(1, 2, figsize=(14, 5))
sns.boxplot(x='cut', y='price', data=diamonds, ax=axs[0])
axs[0].set_title('Original Y-axis')

# Truncated y-axis
axs[1].set_ylim(0, 5000)
sns.boxplot(x='cut', y='price', data=diamonds, ax=axs[1])
axs[1].set_title('Truncated Y-axis')

# Zero-based y-axis
axs[1].set_ylim(0, 5000)
```

_Reflection for Task 5:_
- di plot pertama kita bisa lihat bahwa setiap potongan berlian itu punya harga yang sangat mahal untuk semua potongan berlian, perbedaan harga median (harga tengah) antar jenis potongan terlihat sangat kecil dan tidak signifikan karena skalanya yang besar.
 - di plot kedua bisa kita lihat bahwa median harga potongan 'Ideal' ternyata lebih rendah dari 'Premium' atau bahkan 'Fair'. Perbedaan yang tadinya tidak terlihat kini menjadi sangat jelas.
