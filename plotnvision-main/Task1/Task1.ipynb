## Task 1: Contrast & Color

**Instructions:**
- Create a bar plot showing the average price of diamonds by cut using a low-contrast color scheme.
- Create the same plot using a high-contrast color palette.

**Reflection:**
- Which plot is easier to interpret? Why? Relate your answer to human visual perception.

```python
# Task 1: Your code here
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')
avg_price_by_cut = diamonds.groupby('cut')['price'].mean().reset_index()

# Low-contrast plot
plt.figure(figsize=(10, 6))
sns.barplot(x='cut', y='price', data=avg_price_by_cut, palette='pastel')
plt.title('Average Diamond Price by Cut (Low Contrast)')
plt.xlabel('Cut')
plt.ylabel('Average Price')
plt.show()

# High-contrast plot
plt.figure(figsize=(10, 6))
high_contrast_palette = sns.color_palette("deep")
sns.barplot(x='cut', y='price', data=avg_price_by_cut, palette=high_contrast_palette)
plt.title('Average Diamond Price by Cut (High Contrast)')
plt.xlabel('Cut')
plt.ylabel('Average Price')
plt.grid(axis='y', linestyle='--')
plt.show()
```
